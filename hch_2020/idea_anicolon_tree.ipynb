```python
import pandas as pd

moms = pd.read_csv('moms.csv', usecols=['edad', 'ascendencia',
       'estado_laboral', 'sueldo', 'busca_otro_trabajo', 'actividad',
       'asistencia alguna vez a enseñanza', 'edu preescolar', 'edu primaria',
       'edu media', 'edu técnica', 'edu univ o simil', 'edu terciario no uni',
       'edu posgrado', 'edu mag prof', 'hijos'])

moms['estado_laboral'] = moms['estado_laboral'].astype('category').cat.codes
moms['actividad'] = moms['actividad'].astype('category').cat.codes

moms.columns
```

```python
# Dummies para modelo
 
moms = pd.get_dummies(moms, drop_first=True)

moms
```

```python
# Feautures
X = moms.drop('hijos', axis=1)

# Target
y = moms['hijos']
```

```python
# Split #30% test, 70% para train
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42) 
```

```python
# Create model
 
from sklearn.tree import DecisionTreeClassifier 
tree = DecisionTreeClassifier(max_depth=2, random_state = 42) 
```

```python
# Train model
tree.fit(X_train,y_train)
```

```python
# Exportar el modelo
import joblib
joblib.dump(tree,'tree.pkl')
```

```python
modelo = joblib.load('tree.pkl')
```

```python
# Predicciones
 
y_train_pred = tree.predict(X_train)
y_test_pred = tree.predict(X_test)
```

```python
# Accuracy score

from sklearn.metrics import accuracy_score

train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print('Train accuracy', train_accuracy)
print('Test accuracy',test_accuracy)
```
